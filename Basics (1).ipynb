```python
%%sql
SELECT g.name, 
COUNT(il.track_id), 
ROUND(CAST(count(il.track_id) AS FLOAT) / 
      (SELECT COUNT(track_id) FROM invoice_line
       JOIN invoice USING(invoice_id)
        WHERE billing_country = 'USA') *100, 2) '%'
FROM genre g
JOIN track t USING(genre_id)
JOIN invoice_line il USING(track_id)
JOIN invoice i USING(invoice_id)
WHERE i.billing_country = 'USA'
GROUP BY 1
ORDER BY 2 DESC
```

```python
%%sql
SELECT e.first_name || " " || e.last_name employee_name, 
    ROUND(SUM(i.total)) sales_made
FROM employee e
JOIN customer c
ON e.employee_id = c.support_rep_id
JOIN invoice i USING(customer_id)
GROUP BY 1
ORDER BY 2 DESC
```

```python
%%sql
SELECT e.first_name || " " || e.last_name employee_name, 
       e.hire_date,
       e.title,
       ROUND(SUM(i.total)) sales_made
FROM employee e
JOIN customer c
ON e.employee_id = c.support_rep_id
JOIN invoice i USING(customer_id)
GROUP BY 1
ORDER BY 3 DESC
```

```python
%%sql
SELECT c.country,
       COUNT(DISTINCT c.customer_id) total_num_customers, 
       ROUND(SUM(i.total), 2) total_sales,
       ROUND(CAST (SUM(i.total) AS FLOAT) / COUNT(distinct c.customer_id), 2) avg_sales_per_customer,
       ROUND(AVG(i.total), 2) avg_order_value
FROM customer c
JOIN invoice i USING(customer_id)
GROUP BY 1
ORDER BY 2 DESC
```

```python
%%sql
SELECT a.name,
       COUNT(p.playlist_id) playlist_appearance
FROM artist a
JOIN ALBUM USING(artist_id)
JOIN track USING(album_id)
JOIN playlist_track USING(track_id)
JOIN playlist p USING(playlist_id)
GROUP BY 1
ORDER BY 2 DESC
```

```python
%%sql
SELECT COUNT(DISTINCT t.track_id) total_tracks, 
       COUNT(DISTINCT il.track_id) tracks_purchased,
       COUNT(DISTINCT t.track_id) - 
       COUNT(DISTINCT il.track_id) tracks_notpurchased
    FROM track t
    LEFT JOIN invoice_line il USING(track_id)
    
```

```python
%%sql
WITH media_type_protection as 
(SELECT *,
CASE
    WHEN name LIKE '%prot%' THEN 'Protected'
    ELSE 'Not protected'
    END AS Category
FROM media_type
ORDER BY Category)  
SELECT mtp.name, 
       mtp.category, 
       ROUND(SUM(il.unit_price * il.quantity)) total_sales,
       ROUND(CAST(SUM(il.unit_price * il.quantity) as FLOAT) / 
            (SELECT SUM(invoice_line.unit_price * invoice_line.quantity) FROM invoice_line) * 100)
FROM media_type_protection mtp
JOIN track USING(media_type_id)
JOIN invoice_line il USING(track_id)
GROUP BY 2
ORDER BY 3 DESC
```
